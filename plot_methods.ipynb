```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths = ["D:\\all\\all project\Quanvolution proj\history_send\history_basline_CNN_ver100_fullver.json",
         "D:\\all\\all project\Quanvolution proj\history_send\historyQuantum_Saliency_Mix_ver100_fullver.json",
         "D:\\all\\all project\Quanvolution proj\history_send\historyQuantum_Saliency_Mix_ver100_with_offset.json"]
arg_data = {"baseline_cnn":[],
            "quantum_saliencymix":[],
            "quantum_saliencymix_offset":[]}

for idx, path in enumerate(paths):
    with open(path, 'r') as file:
        data = json.load(file)
        key = list(arg_data.keys())[idx]
        arg_data[key] = data

print (arg_data)
```

```python
def process(info, metric='acc'):
    names = ["baseline_cnn", "quantum_saliencymix", "quantum_saliencymix_offset"]
    color_set = [ "#20C914","#FC980BFF", "#0E8CD4EE"]
    plt.figure(figsize=(12, 5))
    for idx, _ in enumerate(names):

        epochs = np.arange(1, 101)
        plt.plot(epochs, info[_]["val_" + metric], label=_, color = color_set[idx])
        surplus = "loss"
        if metric =='acc':
            surplus = "accuracy"
        plt.title(f"Validation {surplus}")
        plt.xlabel("Epochs")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
    plt.show()
```

```python
process(arg_data)
process(arg_data, "loss")
```
